```python
import numpy as np
import matplotlib.pyplot as plt

def densidade_probabilidade(x, media, desvio_padrao):
    return (1 / (desvio_padrao * np.sqrt(2 * np.pi))) * np.exp(-((x - media) ** 2) / (2 * desvio_padrao ** 2))

# Parâmetros das distribuições normais
medias = [0, 1, -1]  # Diferentes valores de média
desvios_padrao = [1, 0.5, 2]  # Diferentes valores de desvio padrão
cores = ['b', 'g', 'r']  # Cores para os plots

# Vetor de valores no eixo x
valores_x = np.linspace(-5, 5, 1000)

# Criando o gráfico
plt.figure(figsize=(8, 6))

# Plotando a densidade de probabilidade para cada distribuição normal
for media, desvio_padrao, cor in zip(medias, desvios_padrao, cores):
    densidades = densidade_probabilidade(valores_x, media, desvio_padrao)
    plt.plot(valores_x, densidades, label=f'Média={media * qtd_bins}, Desvio Padrão={desvio_padrao}', color=cor)

# Adicionando legendas e título
plt.title('Densidade de Probabilidade de Distribuições Normais')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)
plt.show()
```
